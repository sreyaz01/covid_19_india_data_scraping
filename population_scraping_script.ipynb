## Population Data Scraping

### Source : http://statisticstimes.com/demographics/population-of-indian-states.php

#### Estimated Population of 2018



```python
import requests 
import pandas as pd

# Hide warnings
import warnings
warnings.filterwarnings('ignore')

print('Pandas Version',pd.__version__)
```

```python
link = "http://statisticstimes.com/demographics/population-of-indian-states.php"

web = requests.get(link)
dfs = pd.read_html(web.text)
```

```python
df = dfs[1].copy()
```

```python
df.head()
```

```python
df.columns = [col[1] for col in df.columns]
```

```python
pop2018df = df[['State','2018']].loc[:35].sort_values('2018', ascending = False)

pop2018df
```
